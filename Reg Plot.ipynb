```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.regplot(x="total_bill",y="tip",data=tips)
```

```python
sns.regplot(x="total_bill",y="tip",data=tips,color="green")
```

```python
mean=[2,5]
cov=[[1.1,0.4],[2.2,3]]
```

```python
x_value,y_value=np.random.multivariate_normal(mean,cov,100).T
```

```python
sns.regplot(x=x_value,y=y_value)
```

```python
sns.regplot(x=x_value,y=y_value,color="purple")
```

```python
series_x_value=pd.Series(x_value,name="var-x")
```

```python
series_y_value=pd.Series(y_value,name="var-y")
```

```python
sns.regplot(x=series_x_value,y=series_y_value)
```

```python
sns.regplot(x=series_x_value,y=series_y_value,marker="D")
```

```python
sns.regplot(x="total_bill",y="tip",data=tips,scatter_kws={"color":"red","marker":"o"},line_kws={"color":"green","linewidth":3.1})
```

```python
sns.regplot(x=x_value,y=y_value,ci=34)
```

```python
sns.regplot(x="size",y="total_bill",data=tips,x_jitter=0.5)
```

```python
sns.regplot(x="size",y="total_bill",data=tips,x_estimator=np.mean)
```
