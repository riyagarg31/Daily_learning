```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.stripplot(x=tips["tip"],color="green")
```

```python
sns.stripplot(x=tips["total_bill"])
```

```python
sns.stripplot(x="day",y="total_bill",data=tips)
```

```python
sns.stripplot(x="tip",data=tips)
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,jitter=True)
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,jitter=0.2)
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,hue="sex",jitter=1)
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,hue="sex",jitter=1,linewidth=2)
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,hue="sex",jitter=True,dodge=True,palette="winter_r")
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,hue="smoker",jitter=True,dodge=True,order=["sun","Fri","Thur","Sat"]);
```

```python
sns.stripplot(x="sex",y="tip",data=tips,order=["Female","Male"],palette="coolwarm")
```

```python
sns.stripplot(x="sex",y="tip",data=tips,order=["Male","Female"],marker="D",palette="autumn")
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,marker="D",size=33,hue="sex",edgecolor="green",alpha=0.3)
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,jitter=True,palette="winter_r")
sns.boxplot(x="day",y="total_bill",data=tips)
plt.grid()
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,jitter=True)
sns.violinplot(x="day",y="total_bill",data=tips,color="0.6")
```
