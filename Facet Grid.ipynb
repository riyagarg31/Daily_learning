```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
x=sns.FacetGrid(tips,row="smoker",col="time")
x=x.map(plt.hist,"total_bill")
```

```python
x=sns.FacetGrid(tips,row="smoker",col="time")
x=x.map(plt.hist,"total_bill",color="red",bins=15)
```

```python
x=sns.FacetGrid(tips,row="smoker",col="time")
x=x.map(plt.scatter,"total_bill","tip")
```

```python
x=sns.FacetGrid(tips,row="smoker",col="time")
x=x.map(sns.regplot,"total_bill","tip")
```

```python
x=sns.FacetGrid(tips,hue="smoker",col="time")
x=x.map(plt.scatter,"total_bill","tip")
x.add_legend()
```

```python
x=sns.FacetGrid(tips,col="day",height=7)
x=x.map(sns.boxplot,"total_bill","time")
```

```python
x=sns.FacetGrid(tips,col="day",height=7,col_order=["Thur","Fri","Sun","Sat"])
x=x.map(sns.boxplot,"total_bill",color="green")
```

```python
x=sns.FacetGrid(tips,col="day",height=7,col_order=["Thur","Fri","Sun","Sat"],palette="husl")
x=x.map(sns.boxplot,"total_bill")
```
