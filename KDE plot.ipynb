```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mean=[0,0]
```

```python
cov=[[0.2,0],[0,3]]
```

```python
x_axis,y_axis=np.random.multivariate_normal(mean,cov,size=40).T
```

```python
sns.kdeplot(x_axis,shade=True,color="red")
```

```python
sns.kdeplot(y_axis,shade=True,color="purple")
```

```python
sns.kdeplot(x=x_axis,y=y_axis)
```

```python
sns.kdeplot(x=x_axis,y=y_axis,shade=True,cmap="coolwarm")
```

```python
sns.kdeplot(x=x_axis,y=y_axis,shade=True,cmap="RdBu",n_levels=18)
```

```python
sns.kdeplot(y=y_axis)
```
