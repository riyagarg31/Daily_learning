```python
import pandas as pd
```

```python
pd.set_option("display.max_rows",None,"display.max_columns",None)
```

```python
fortune=pd.read_csv("fortune1000.csv",index_col="Rank")
fortune
```

```python
sectors=fortune.groupby("Sector")
```

```python
sectors["Profits"].max()
```

```python
df=pd.DataFrame(columns=fortune.columns)
df
```

```python
for sector, data in sectors:
    highest_revenue=data.nlargest(1,"Revenue")
    df=df.append(highest_revenue)
```

```python
df
```

```python
cities=fortune.groupby("Location")
```

```python
df=pd.DataFrame(columns=fortune.columns)
```

```python
df
```

```python
for location, data in cities:
    highest_revenue_city=data.nlargest(1,"Revenue")
    df=df.append(highest_revenue_city)
```
