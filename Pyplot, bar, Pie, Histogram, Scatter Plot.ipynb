```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
month=["jan", "feb", "mar", "apr", "may"]
HYD_sales = [50, 60, 55, 75, 65]
BANG_sales = [55, 50, 70, 60, 80]
index=np.arange(1,6)
width=0.30
plt.bar(index,HYD_sales, width, color="green",label="HYD SALES")
plt.bar(index+width, BANG_sales, width, color="blue",label="BANG SALES")
plt.title("HORIZONTALLY STACKED BAR GRAPH")
plt.xlabel("MONTH")
plt.ylabel("SALES")
plt.xticks(index+width/2,month)
plt.legend(loc="best")
plt.show()
```

```python
month=["aug","sep","oct","nov","dec"]
sales1=[50,60,55,75,65]
sales2=[55,50,70,60,80]
index=np.arange(1,6)
width=0.30
plt.barh(index,sales1,width,color="red",label="SALES1")
plt.barh(index+width,sales2,width,color="green",label="SALES2")
plt.title("SALES")
plt.xlabel("MONTH")
plt.ylabel("SALES")
plt.xticks(index+width/2,month)
plt.legend(loc="upper right")
plt.show()
```

### PIE CHART

```python
month=["jan","feb","mar","apr","may"]
HYD_sales=[50,60,55,75,65]
plt.pie(HYD_sales,explode=[0,0,0.1,1,0.1],labels=month,shadow=True)
plt.title("SALES GRAPH")
plt.legend()
plt.show()
```

### Histogram

```python
a=np.random.randn(1000)
plt.hist(a,10)
plt.title("HISTOGRAM")
plt.xlabel("DATES")
plt.ylabel("FREQUENCY")
plt.grid()
plt.show()
```

### Scatter Plot

```python
a=np.random.randn(100)
b=np.random.randn(100)
plt.scatter(a,b)
plt.title("SCATTER PLOT")
plt.grid()
plt.show()
```
