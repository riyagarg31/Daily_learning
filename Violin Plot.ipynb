```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.violinplot(x=tips["total_bill"])
```

```python
sns.violinplot(x="size",y="total_bill",data=tips)
```

```python
sns.violinplot(x="day",y="total_bill",data=tips)
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="sex",palette="autumn")
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="smoker",palette="husl")
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="smoker",palette="coolwarm",split=True,order=["Sun","Thur","Sat","Fri"])
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="sex",palette="husl",inner="quartile",split=True)
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="smoker",inner="stick",scale="count")
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="sex",palette="RdBu",bw=0.3,scale_hue=False)
```
