### Multiple Plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
stocks_df=pd.read_csv("stocks.csv")
```

```python
stocks_df
```

```python
stocks_df.plot(x="Date",y=["AAPL","sp500"],linewidth=5)
```

```python
stocks_df.plot(x="Date",y=["AAPL","sp500"],linewidth=5)
plt.ylabel("Price")
plt.title("Stock Prices")
plt.grid()
```

```python
stocks_df.plot(x="Date",y=["AAPL","sp500","GOOG"],linewidth=3)
plt.ylabel("Price")
plt.title("Stock Prices")
plt.legend(loc="upper center")
plt.grid()
```

```python
stocks_df.plot(x="Date",y=["AAPL","sp500","GOOG"],linewidth=3)
plt.ylabel("Price")
plt.title("Stock Prices")
plt.legend(loc="upper center")
plt.grid()
```

### Sub Plots

```python
plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
plt.plot(stocks_df["AAPL"],"r--")
plt.grid()
plt.subplot(1,2,2)
plt.plot(stocks_df["sp500"],"b.")
plt.grid()
```

```python
plt.figure(figsize=(8,8))
plt.subplot(2,1,1)
plt.plot(stocks_df["AAPL"],"r--")
plt.grid()
plt.subplot(2,1,2)
plt.plot(stocks_df["sp500"],"b.")
plt.grid()
```

```python
plt.figure(figsize=(8,8))
plt.subplot(3,1,1)
plt.plot(stocks_df["AAPL"],"r--")
plt.grid()
plt.legend(["AAPLE price"])
plt.subplot(3,1,2)
plt.plot(stocks_df["sp500"],"b.")
plt.grid()
plt.legend(["S&P500 price"])
plt.subplot(3,1,3)
plt.plot(stocks_df["GOOG"],"y--")
plt.grid()
plt.legend(["GOOGLE price"])
```
