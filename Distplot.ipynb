```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
num=np.random.randn(150)
```

```python
sns.distplot(num,color="red")
plt.grid()
```

```python
label_dist=pd.Series(num,name="variablex")
```

```python
sns.distplot(label_dist)
```

```python
sns.distplot(label_dist,vertical=True,hist=False)
```

```python
sns.distplot(label_dist,rug=True,hist=False,color="green")
```
