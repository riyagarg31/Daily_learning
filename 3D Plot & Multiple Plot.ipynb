### Multiple Plots

```python
plt.subplot(1,2,1)
plt.plot([1,2,3,4],[10,20,30,40])
plt.title("1st PLOT")
plt.subplot(1,2,2)
x=np.arange(1,5)
y=x**2
plt.plot(x,y)
plt.title("2nd PLOT")
plt.suptitle("SALES PLOT")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.scatter([10,20,30,40,50],[100,200,300,400,500])
plt.title("1st PLOT")
plt.subplot(2,1,2)
plt.bar([1,2,3,4,5],[10,20,30,40,50])
plt.title("2nd PLOT")
plt.suptitle("PLOTS")
plt.show()
```
