```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=sns.load_dataset("iris")
```

```python
iris
```

```python
x=sns.PairGrid(iris)
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris)
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(iris,hue="species")
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(iris,hue="species")
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
x=x.add_legend(loc="upper right")
```

```python
x=sns.PairGrid(iris,hue="species",palette="coolwarm")
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue="species",palette="husl")
x=x.map_diag(plt.hist,histtype="step",linewidth=5)
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(iris,vars=["petal_width","petal_length"])
x=x.map_diag(plt.hist,edgecolor="black")
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(iris,x_vars=["petal_length","petal_width"],y_vars=["sepal_length","sepal_width"])
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris)
x=x.map_diag(plt.hist,edgecolor="red")
x=x.map_upper(plt.scatter)
x=x.map_lower(sns.kdeplot)
```
