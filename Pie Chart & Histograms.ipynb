```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Pie Chart

```python
values=[15,55,17,4,9]
colors=["r","g","b","y","m"]
labels=["AAPL","GOOG","T","TSLA","AMZN"]
plt.figure(figsize=(7,7))
plt.pie(values,colors=colors,labels=labels)
```

```python
values=[15,55,17,4,9]
colors=["r","g","b","y","m"]
labels=["AAPL","GOOG","T","TSLA","AMZN"]
explode=[0,0.3,0,0,0.3]
plt.figure(figsize=(7,7))
plt.pie(values,colors=colors,labels=labels,explode=explode);
plt.title("STOCK PORTFOLIO")
```

### Histogram

```python
daily_return_df=pd.read_csv("daily_returns.csv")
```

```python
daily_return_df
```

```python
mu=daily_return_df["AAPL"].mean()
```

```python
sigma=daily_return_df["AAPL"].std()
```

```python
mu
```

```python
sigma
```

```python
num_bins=40
plt.figure(figsize=(7,5))
plt.hist(daily_return_df["AAPL"],num_bins);
plt.grid()
plt.title("Histogram:mu="+str(mu)+",sigma="+str(sigma))
```
