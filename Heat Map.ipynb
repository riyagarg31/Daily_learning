```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
normal=np.random.rand(12,15)
```

```python
sns.heatmap(normal)
```

```python
sns.heatmap(normal,linewidth=3)
```

```python
sns.heatmap(normal,annot=True)
```

```python
sns.heatmap(normal,vmin=0,vmax=2)
```

```python
flights=sns.load_dataset("flights")
```

```python
flights
```

```python
flights=flights.pivot("year","month","passengers")
```

```python
sns.heatmap(flights)
```

```python
sns.heatmap(flights,annot=True)
```

```python
sns.heatmap(flights,annot=True,fmt="d")
```

```python
sns.heatmap(flights,annot=True,fmt="d",cmap="autumn")
```

```python
sns.heatmap(flights,cbar=False)
```

```python
sns.heatmap(flights,center=flights.loc[1954,"June"],annot=True,fmt="d")
```
